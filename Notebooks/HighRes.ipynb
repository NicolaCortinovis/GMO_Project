## High resolution target image

This notebook contains the results for individuals generated towards a low resolution target image

```python
import sys
import os
import matplotlib.pyplot as plt
from PIL import Image

GA = os.path.abspath(os.path.join(os.getcwd(), '..', 'GA'))

sys.path.append(GA)

image_path = 'Images/highres/birds.png'  # 1920x1080

from evolution import genetic_algorithm
```

```python
img = Image.open(image_path)
plt.imshow(img)
plt.title('Original Image: 1920 x 1080 pixels')
plt.axis('off')
plt.show()
```

```python
result, fitess_evolution = genetic_algorithm(
    image_path = image_path,
    factor = 10,
    pop_size=50,
    generations=20000,
    mutation_rate=0.05,
    tournament_size=5,
    replacement_size = 15,
    c = 1.1,
    max_mut_rate = 0.35
)
```

```python
plt.plot(fitess_evolution)
plt.xlabel('Generation')
plt.ylabel('Fitness')
plt.title('Fitness Evolution for high resolution image')
plt.show()
```
